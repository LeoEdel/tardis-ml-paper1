## Get lat lon

```python
gsfc_file = '/scratch/project_465000269/edelleo1/Leo/SIT_observations/ICESat1/tif∕gsfc_25n.tif'
lon_file = '/scratch/project_465000269/edelleo1/Leo/SIT_observations/ICESat1/tif∕PS25km_north_lon.tif'
lat_file = '/scratch/project_465000269/edelleo1/Leo/SIT_observations/ICESat1/tif∕PS25km_north_lat.tif'
```

```python
lat_is = np.array(Image.open(lat_file))
lon_is = np.array(Image.open(lon_file))
gsfc_is = np.array(Image.open(gsfc_file))
```

```python
plt.imshow(gsfc_is)
plt.colorbar()
```

# Clean the data

```python
sit = xr.DataArray(
    data=sit_is,
    dims=["time", "y", "x"],
    coords=dict(
        time=(["time"], dates_avg),
        x=(["x"], range(304)),
        y=(["y"], range(448)[::-1]),
        lon=(["y", "x"], lon_is),
        lat=(["y", "x"], lat_is)
    ),
    attrs=dict(
        name="Sea ice thickness",
        description="ICESat 1",
        project='TARDIS',
        units="m",
    ),
)

sit.name = 'sit'
```

```python
sit = sit.where(gsfc_is<=0)  # put nan where land
```

```python
sit = sit.where(sit>=-0.5)  # put nan where open ocean
```

```python
sit.isel(time=10).plot(vmax=4)
```

```python
odir = '/scratch/project_465000269/edelleo1/Leo/SIT_observations/ICESat1/nc/'
ofile = 'SIT_gridded_ICESat1_campaigns_co_Envisat.nc'
```

```python
sit.to_netcdf(f'{odir}{ofile}')
print(f'ICESat1 gridded saved as: {odir}{ofile}')
```
