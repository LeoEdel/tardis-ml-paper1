```python
def draw_all_bp(models, days, rootdir, fig_dir, showfig=False, savefig=True, cs2smos=None, vmax=4):
    for day in days:
        draw_bp(models, day, rootdir=rootdir, fig_dir=fig_dir, savefig=True, showfig=showfig, cs2smos=cs2smos, vmax=vmax)
```

```python
def draw_all_bps(models, days, rootdir, fig_dir, showfig=False, savefig=True, vmax=4):
    for day in days:
        draw_bp_simple(models, day, rootdir=rootdir, fig_dir=fig_dir, savefig=True, showfig=showfig, vmax=vmax)
```

```python
def draw_bp_simple(models, day, rootdir='', fig_dir='', showfig=False, savefig=False, vmax=4):
    '''Plot spatial reconstruction
    
    Parameters:
    -----------
        models     :     array containing multiple class SITCorrected
        day        :     datetime.datetime object, day to plot
        
        cs2smos    :     xarray DataArray, sea ice thickness merged product from CS2SMOS
                            will be plotted if not None
        vmax       :     int, threshold maximum for SIT. 4 is default and optimal for 2010-2020, 5/6 may be optimal for 2000-2010
    '''
    
    fig, axes = plt.subplots(ncols=3, nrows=2, figsize=(28,18), constrained_layout=True)

    # first model in models will be usd to plot default variables
    model_default = models[list(models.keys())[0]] 

    # identify index to plot
    chrono_dt = np.array([dt.date() for dt in model_default.chrono.date])
    idx = np.where(chrono_dt==day.date())[0]
    
    if len(idx) < 1:
        print('Day not found')
        return
    
    # contour plot
    levels = np.arange(1, vmax+1, 1)
    
    # plot TOPAZ4b
    model_default.sit_a.isel(time=idx).plot(ax=axes[0][0], vmin=0, vmax=vmax, add_colorbar=False)  # TOPAZ4b    
        # contour need an int
    cl = model_default.sit_a.isel(time=idx[0]).plot.contour(ax=axes[0][0], levels=levels, vmin=0, vmax=vmax, add_colorbar=False, cmap=plt.get_cmap('Greys'))
        
    
#     if not bias:  # print SIT for each model
    cmap_extend = 'max'
    model_default.sit_bl.isel(time=idx).plot(ax=axes[1][1], vmin=0, vmax=vmax, add_colorbar=True, 
                                             extend='max', cbar_kwargs={'orientation':'horizontal',
                                                                       'label':'SIT (m)',
                                                                       'ticks':[0,1,2,3,4]})

    cl = model_default.sit_bl.isel(time=idx[0]).plot.contour(ax=axes[1][1], levels=levels, vmin=0, vmax=vmax, add_colorbar=False, cmap=plt.get_cmap('Greys'))



    model_default.sit.isel(time=idx).plot(ax=axes[0][1], vmin=0, vmax=vmax, add_colorbar=False)
    cl = model_default.sit.isel(time=idx[0]).plot.contour(ax=axes[0][1], levels=levels, vmin=0, vmax=vmax, add_colorbar=False, cmap=plt.get_cmap('Greys'))
    axes[0][1].set_title(f'LSTM')
    



    # bias
    import matplotlib.colors as colors
     # diverent colormap [black - white - black]
    colors_under = plt.get_cmap('Greys_r')(np.linspace(0, 1, 256))
    colors_over = plt.get_cmap('Greys')(np.linspace(0, 1, 256))
    all_colors = np.vstack((colors_under, colors_over))
    contourmap = colors.LinearSegmentedColormap.from_list('mymap', all_colors)
    
    
    cmap = plt.get_cmap('coolwarm_r')
    bias = model_default.sit.isel(time=idx) - model_default.sit_a.isel(time=idx)
    bias.plot(ax=axes[0][2], vmin=-2, vmax=2, add_colorbar=False, cmap=cmap)
    
    levels = np.arange(-2, 2.5, 0.5)
    bias.isel(time=0).plot.contour(ax=axes[0][2], levels=levels, vmin=-2, vmax=2, add_colorbar=False, cmap=contourmap)

    bias_bl = model_default.sit_bl.isel(time=idx) - model_default.sit_a.isel(time=idx)
    bias_bl.plot(ax=axes[1][2], vmin=-2, vmax=2, add_colorbar=True, cmap=cmap, 
              extend='both', cbar_kwargs={'orientation':'horizontal',
                                         'label':'SIT bias (m)',
                                         'ticks':[-2,-1,0,1,2]})
    
    bias_bl.isel(time=0).plot.contour(ax=axes[1][2], levels=levels, vmin=-2, vmax=2, add_colorbar=False, cmap=contourmap)
    
    

    axes[0][2].set_title(f'LSTM - [TOPAZ+CS2SMOS]')
    axes[1][2].set_title(f'Baseline - [TOPAZ+CS2SMOS]')
    
    axes[0][0].set_title(f'TOPAZ+CS2SMOS') # TOPAZ4b
    axes[1][1].set_title(f'Baseline')

    axes[1][0].set_visible(False)
        

    for ax in axes.flatten():
        ax.set_xlabel('')
        ax.set_ylabel('')
        ax.tick_params(labelbottom=False, labelleft=False, bottom=False, left=False)
        
    
    fig.suptitle(f'{chrono_dt[idx][0].strftime("%Y %m %d")}', fontsize=30)

    modif_plot.resize(fig, s=30)
    
    if savefig:
        sdate = chrono_dt[idx][0].strftime("%Y%m%d")
        ofile = f'{rootdir}{fig_dir}SIT_LSTM_bs_beautyplots_{sdate}.png'
            
        plt.savefig(f"{ofile}", dpi=124, facecolor='white')
        print(f'Saved as : {ofile}')
    
    if showfig:
        plt.show()
    
    plt.close()        
```

```python
def draw_bp(models, day, rootdir='', fig_dir='', showfig=False, savefig=False, bias=False, apply=False, cs2smos=None, vmax=4):
    '''Plot spatial reconstruction
    
    Parameters:
    -----------
        models     :     array containing multiple class SITCorrected
        day        :     datetime.datetime object, day to plot
        
        bias       :     bool, if False (default) will plot SIT. if True, will plot SIT (TOPAZ4b - model) for model in models
        apply      :     bool, if True: will hide TOPAZ4b (do not exist) nor Baseline (not done yet)
        
        cs2smos    :     xarray DataArray, sea ice thickness merged product from CS2SMOS
                            will be plotted if not None
        vmax       :     int, threshold maximum for SIT. 4 is default and optimal for 2010-2020, 5/6 may be optimal for 2000-2010
    '''
    
    fig, axes = plt.subplots(ncols=3, nrows=2, figsize=(28,18), constrained_layout=True)

    # first model in models will be usd to plot default variables
    model_default = models[list(models.keys())[0]] 

    # identify index to plot
    chrono_dt = np.array([dt.date() for dt in model_default.chrono.date])
    idx = np.where(chrono_dt==day.date())[0]
    
    if len(idx) < 1:
        print('Day not found')
        return
    
    # contour plot
    levels = np.arange(1, vmax+1, 1)
    
    # plot TOPAZ4b
    model_default.sit_a.isel(time=idx).plot(ax=axes[0][1], vmin=0, vmax=vmax, add_colorbar=False)  # TOPAZ4b    
        # contour need an int
    cl = model_default.sit_a.isel(time=idx[0]).plot.contour(ax=axes[0][1], levels=levels, vmin=0, vmax=vmax, add_colorbar=False, cmap=plt.get_cmap('Greys'))
        
    
#     if not bias:  # print SIT for each model
    cmap_extend = 'max'
    imC = model_default.sit_na.isel(time=idx).plot(ax=axes[0][0], vmin=0, vmax=vmax, add_colorbar=False)  # label='TP4 Free Run', c='#1f77b4')
    model_default.sit_bl.isel(time=idx).plot(ax=axes[1][0], vmin=0, vmax=vmax, add_colorbar=True, 
                                             extend='max', cbar_kwargs={'orientation':'horizontal',
                                                                       'label':'SIT (m)',
                                                                       'ticks':[0,1,2,3,4]})

    cl = model_default.sit_na.isel(time=idx[0]).plot.contour(ax=axes[0][0], levels=levels, vmin=0, vmax=vmax, add_colorbar=False, cmap=plt.get_cmap('Greys'))
    cl = model_default.sit_bl.isel(time=idx[0]).plot.contour(ax=axes[1][0], levels=levels, vmin=0, vmax=vmax, add_colorbar=False, cmap=plt.get_cmap('Greys'))



    model_default.sit.isel(time=idx).plot(ax=axes[1][1], vmin=0, vmax=vmax, add_colorbar=True, 
                                          extend='max', cbar_kwargs={'orientation':'horizontal',
                                                                    'label':'SIT (m)',
                                                                    'ticks':[0,1,2,3,4]})
    cl = model_default.sit.isel(time=idx[0]).plot.contour(ax=axes[1][1], levels=levels, vmin=0, vmax=vmax, add_colorbar=False, cmap=plt.get_cmap('Greys'))
#     axes[1][1].set_title(f'{model_default.name}')
    axes[1][1].set_title(f'LSTM')
    



    # bias
    import matplotlib.colors as colors
     # diverent colormap [black - white - black]
    colors_under = plt.get_cmap('Greys_r')(np.linspace(0, 1, 256))
    colors_over = plt.get_cmap('Greys')(np.linspace(0, 1, 256))
    all_colors = np.vstack((colors_under, colors_over))
    contourmap = colors.LinearSegmentedColormap.from_list('mymap', all_colors)
    
    
    cmap = plt.get_cmap('coolwarm_r')
    bias = model_default.sit.isel(time=idx) - model_default.sit_a.isel(time=idx)
    bias.plot(ax=axes[1][2], vmin=-2, vmax=2, add_colorbar=True, cmap=cmap, 
              extend='both', cbar_kwargs={'orientation':'horizontal',
                                         'label':'SIT bias (m)',
                                         'ticks':[-2,-1,0,1,2]})
    
    levels = np.arange(-2, 2.5, 0.5)
    cl = bias.isel(time=0).plot.contour(ax=axes[1][2], levels=levels, vmin=-2, vmax=2, add_colorbar=False, cmap=contourmap)

#     axes[1][2].set_title(f'{model_default.name} - TP4b')
    axes[1][2].set_title(f'LSTM - [TOPAZ+CS2SMOS]')
    
            


    axes[0][1].set_title(f'TOPAZ+CS2SMOS') # TOPAZ4b
    axes[0][0].set_title(f'TOPAZ')  # TOPAZ4b FreeRun
    axes[1][0].set_title(f'Baseline')
    
    
    # if we got observation data from CS2SMOS:
    # we plot
    # weekly mean on each day (average including +- 3 days)
    if cs2smos is not None:
        # identify index to plot
        cs_chrono = pd.DataFrame({'date':pd.to_datetime(cs2smos['time'].to_numpy())})
        cs_chrono_dt = np.array([dt.date() for dt in cs_chrono.date])
        cs_idx = np.where(cs_chrono_dt==day.date())[0]
        # plot if day exists (no data in summer)
        if len(cs_idx) == 1:
            cs2smos.isel(time=cs_idx, yc=slice(90,310), xc=slice(100,332)).plot(ax=axes[0][2], vmin=0, vmax=vmax, add_colorbar=False)
            # contour plot
            levels = np.arange(1, 5, 1)
            # contour need an int
            cl = cs2smos.isel(time=cs_idx[0]).plot.contour(ax=axes[0][2], levels=levels, vmin=0, vmax=vmax, add_colorbar=False, cmap=plt.get_cmap('Greys'))
            axes[0][2].set_title(f'CS2SMOS')
            # axes[1][0].clabel(cl, fontsize=18, inline=1)
        else:
            axes[0][2].set_visible(False)
    else:
        axes[0][2].set_visible(False)
    
    
    if apply:
        axes[0][0].set_visible(False)  # TOPAZ4b assimilation does not exist
        

    for ax in axes.flatten():
        ax.set_xlabel('')
        ax.set_ylabel('')
        ax.tick_params(labelbottom=False, labelleft=False, bottom=False, left=False)
        
    
    fig.suptitle(f'{chrono_dt[idx][0].strftime("%Y %m %d")}', fontsize=30)

    modif_plot.resize(fig, s=30)
    
    if savefig:
        sdate = chrono_dt[idx][0].strftime("%Y%m%d")
        ofile = f'{rootdir}{fig_dir}SIT_LSTM3_bk_bp_{sdate}.png'
            
        plt.savefig(f"{ofile}", dpi=124, facecolor='white')
        print(f'Saved as : {ofile}')
    
    if showfig:
        plt.show()
    
    plt.close()        
```
