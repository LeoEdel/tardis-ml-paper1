# Check centered running mean on forcings

```python
forcings = dict()
for field in forcing_fields:
    forcings[field] = np.load(os.path.join(rootdir, forcing_bdir, f'{field}.npy'))
```

```python
# %matplotlib notebook  # for interactive plot
```

```python
fig, ax = plt.subplots(ncols=1, nrows=1,figsize=(16, 9))
    
for field in forcing_fields[::-1]:
    forcings[field].shape
    ax.plot(np.mean(forcings[field], axis=(1,2)), label=f'{field}')
    ax.set_ylabel(f'forcing')
plt.legend();

# plt.savefig("fig/forcing_comparison_{}d.png".format(ndays))
```
