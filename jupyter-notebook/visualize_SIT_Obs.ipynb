### Load

```python
odir = '/nird/projects/nird/NS2993K/Leo/SIT_observations/unified_processed/'
ofile = f'CryoSat-AWI_summaries_SIT_monthlymean.npy'

SIT_cs = np.load(f'{odir}{ofile}')
```

```python
odir = '/nird/projects/nird/NS2993K/Leo/SIT_observations/unified_processed/'
ofile = f'IceBridge-V2_summaries_SIT_monthlymean.npy'

SIT_v2 = np.load(f'{odir}{ofile}')
```

```python
odir = '/nird/projects/nird/NS2993K/Leo/SIT_observations/unified_processed/'
ofile = f'IceBridge-QL_summaries_SIT_monthlymean.npy'

SIT_ql = np.load(f'{odir}{ofile}')
```

----

### Plot

```python
month_labels = ['Dec']+['Mar','Jun','Sept','Dec']*6 # + ['Mar','Jun']
year_labels = ['2010','2011','2012','2013','2014','2015','2016']
```

```python
fig, ax = plt.subplots(figsize=(21,6))

plt.plot(SIT_cs, lw = 3, label='CryoSat2')
plt.plot(SIT_v2, 'v', ms=15, label='IceBridge V2')
plt.plot(SIT_ql, '*', ms=15, label='IceBridge QL')

plt.ylabel('monthly mean SIT (m)')
plt.legend(fontsize=15)

ax.set_xlim([0,74])  #80])

ax.set_xticks(np.arange(0,74,3))
ax.set_xticklabels(month_labels);
ax.grid(True)

ax2 = ax.twiny()
ax2.set_xlim([0,74])

ax2.set_xticks(np.arange(1,74,12));
ax2.set_xticklabels(year_labels);
# Move twinned axis ticks and label from top to bottom
ax2.xaxis.set_ticks_position("bottom")
ax2.xaxis.set_label_position("bottom")
#  Offset the twin axis below the host
ax2.spines["bottom"].set_position(("axes", -0.15))

modif_plot.resize(fig, 22)
fig.tight_layout()

if True:  # savefig:
    odir = '/nird/projects/nird/NS2993K/Leo/SIT_observations/unified_processed/'
    filename = f'MultiObservations_summaries_SIT_monthlymean.png'
    plt.savefig(f"{odir}{filename}")
    print(f'Figure saved as: {odir}{filename}')
```
